```python
# notebooks/eda_frosh.ipynb
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os
sys.path.append(os.path.abspath('../scripts'))
from clean_data import load_and_clean_frosh_app_counts

# Loading and cleaning data
df = load_and_clean_frosh_app_counts()
df.head()

# Plot: Total Applicants by Year
df.groupby('Fall term')['Applicants'].sum().plot(
    kind='line', marker='o', title='Total Frosh Applicants by Year'
)
plt.ylabel('Applicants')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot: Applicants by Residency Type
df.groupby(['Fall term', 'Applicant characteristics'])['Applicants'].sum().unstack().plot(
    kind='line', marker='o', figsize=(10, 6), title='Applicants by Residency Type Over Time'
)
plt.ylabel('Applicants')
plt.grid(True)
plt.tight_layout()
plt.show()
```
